```python
def add_company(input):
    if input in ['Samsung', 'Razer', 'Microsoft', 'Vero', 'Xiaomi',
                   'Chuwi', 'Fujitsu', 'Google', 'LG', 'Huawei']:
        return 'other'
    else:
        return input

data['Company'] = data['Company'].apply(add_company)
    
```

```python
data['cpu_name']=data['Cpu'].apply(lambda x:" ".join(x.split()[0:3]))
```

```python
def set_processor(name):
    name = name.upper()  # optional: makes it case-insensitive
    if 'INTEL CORE I7' in name:
        return 'Intel Core i7'
    elif 'INTEL CORE I5' in name:
        return 'Intel Core i5'
    elif 'INTEL CORE I3' in name:
        return 'Intel Core i3'
    elif 'AMD' in name:
        return 'AMD'
    else:
        return 'other'

data['cpu_name'] = data['Cpu'].apply(set_processor)
```

```python
def set_os(name):
    if 'Windows 10' in name:
        return 'Windows'
    elif 'Windows 7' in name:
        return 'Windows'
    elif 'Windows 10 S' in name:
        return 'Windows'
    elif 'macOS' in name:
        return 'mac'
    elif 'Mac OS X' in name:
        return 'mac'
    elif 'Linux' in name:
        return name
    else:
        return 'other'

data['OpSys'] = data['OpSys'].apply(set_os)

```

```python
data=data.drop(columns=['laptop_ID','Inches','Product','ScreenResolution','Cpu','Gpu'])
```

```python
data.head(2)
```

```python
data = pd.get_dummies(data)
```

```python
data.head(2)
```

```python
data.shape
```

```python
x= data.drop('Price_euros',axis=1)
y=data['Price_euros']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
```

```python
x_train.shape,x_test.shape
```

```python
def model_acc(model):
    model.fit(x_train,y_train)
    acc=model.score(x_test,y_test)
    print(str(model)+ ' -->' +str(acc))
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
model_acc(lr)
from sklearn.linear_model import Lasso
lasso=Lasso()
model_acc(lasso)
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
model_acc(dt)
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
model_acc(rf)
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {
    'n_estimators': [10, 50, 100],
    'criterion': ['squared_error', 'absolute_error', 'poisson']
}

grid_obj = GridSearchCV(estimator=rf, param_grid=parameters)
grid_fit = grid_obj.fit(x_train, y_train)  # make sure your features are X_train
best_model=grid_fit.best_estimator_
best_model
```

```python
best_model.score(x_test,y_test)
```

```python
import pickle
with open('predictor.pickle','wb') as file:
    pickle.dump(best_model,file)
```
